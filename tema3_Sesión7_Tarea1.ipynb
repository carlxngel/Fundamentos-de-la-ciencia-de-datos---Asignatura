#### 3. Tabla de Frecuencias Absolutas y Relativas

```r
# Tabla de frecuencias para Región
cat("=== REGIÓN ===\n")
tabla_region <- table(datos$Region)
freq_region <- data.frame(
    Frecuencia_Absoluta = as.vector(tabla_region),
    Frecuencia_Relativa = round(prop.table(tabla_region), 3),
    Porcentaje = paste0(round(prop.table(tabla_region) * 100, 1), "%")
)
rownames(freq_region) <- names(tabla_region)
print(freq_region)

cat("\n=== SATISFACCIÓN ===\n")
tabla_satisfaccion <- table(datos$Satisfaccion)
freq_satisfaccion <- data.frame(
    Frecuencia_Absoluta = as.vector(tabla_satisfaccion),
    Frecuencia_Relativa = round(prop.table(tabla_satisfaccion), 3),
    Porcentaje = paste0(round(prop.table(tabla_satisfaccion) * 100, 1), "%")
)
rownames(freq_satisfaccion) <- names(tabla_satisfaccion)
print(freq_satisfaccion)

cat("\n=== EDAD ===\n")
tabla_edad <- table(datos$Edad)
freq_edad <- data.frame(
    Frecuencia_Absoluta = as.vector(tabla_edad),
    Frecuencia_Relativa = round(prop.table(tabla_edad), 3),
    Porcentaje = paste0(round(prop.table(tabla_edad) * 100, 1), "%")
)
rownames(freq_edad) <- names(tabla_edad)
print(freq_edad)
```

#### 4. Media y Mediana de la Variable Edad y Satisfacción

**CONSIDERACIONES**
- Ambas variables tienen valores NA que deben tratarse
- Se usará na.rm = TRUE para excluir NAs del cálculo
- Para Satisfacción (ordinal), la mediana es más apropiada que la media
- Los cálculos se basan en n=14 observaciones válidas\n

```r
# Media y Mediana de Edad
cat("=== EDAD ===\n")
cat("Media:", mean(datos$Edad, na.rm = TRUE), "\n")
cat("Mediana:", median(datos$Edad, na.rm = TRUE), "\n")

# Media y Mediana de Satisfacción
cat("\n=== SATISFACCIÓN ===\n")
cat("Media:", mean(datos$Satisfaccion, na.rm = TRUE), "\n")
cat("Mediana:", median(datos$Satisfaccion, na.rm = TRUE), "\n")
```

**Interpretación**

La **edad** promedio de los clientes es 38 años

- La media > mediana indica presencia de edades más altas que elevan el promedio (distribución asimétrica positiva)

La **satisfacción** típica es de 4 (mediana)

- En escalas ordinales, la MEDIANA es la medida más apropiada
- Satisfacción general: BUENA (4 de 5 puntos)

#### 5. Varianza y Desviación Estándar de la Edad

```r
# Varianza y Desviación Estándar de Edad
cat("\n=== EDAD ===\n")
cat("Varianza:", var(datos$Edad, na.rm = TRUE), "\n")
cat("Desviación Estándar:", sd(datos$Edad, na.rm = TRUE), "\n")
# Coeficiente de Variación de Edad
cat("\n=== EDAD ===\n")
cat("Coeficiente de Variación:", 
    (sd(datos$Edad, na.rm = TRUE) / mean(datos$Edad, na.rm = TRUE)) * 100, 
    "%\n")
```

**CONCLUSIONES**

- Desviación estándar: 13.28 años (ALTA dispersión)

- Coeficiente de variación: 34.06% (heterogeneidad marcada)

Interpretación: La empresa atiende a clientes muy diversos en edad, desde jóvenes hasta adultos mayores por lo que requiere estrategias de servicio diferenciadas por grupo etario
